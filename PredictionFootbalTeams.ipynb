```python
# Installing pandas library

import pandas as pd

# Reading in the initial data

matches = pd.read_csv("https://raw.githubusercontent.com/dataquestio/project-walkthroughs/master/football_matches/matches.csv", index_col=0)
```

```python
#Convert data into datetime type (Algoithm can't work with data that is not numeric)

matches["date"] = pd.to_datetime(matches["date"])
matches["venue_code"] = matches["venue"].astype("category").cat.codes
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
matches["day_code"] = matches["date"].dt.dayofweek
matches["result"] = (matches["result"] == "w").astype("int")
```

```python
#Updated data

matches
```

```python
#Setting up a target for machine learning model. The target is whether the team win or not

matches["target"] = (matches["result"] == "W").astype("int")
```

```python
#Importing Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
```

```python
#Initialazing RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
#Creating date type data for training

train = matches[matches["date"] < '2022-01-01']
```

```python
#Declaring the test dataset

test = matches[matches["date"] > '2022-01-01']
```

```python
#Creating a variable with a list with predictors

predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
#Fitting Random Forest model for predicting the target using trained predictors 

rf.fit(train[predictors], train["target"])
```

```python
#Generating predictions using predict methods

preds = rf.predict(test[predictors])
```

```python
#Import accuracy_score what would tell what percentage of the time prediction is accurate

from sklearn.metrics import accuracy_score
```

```python
error = accuracy_score(test["target"], preds)
```

```python
#Looking for an accuracy

error
```

```python
#Creating a DataFrame which will combine actual and predicted values

combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))
```

```python
#Creating a two-way table that will show when predicted zero or one actually happened 

pd.crosstab(index=combined["actual"], columns=combined["predicted"])
```

```python
#Import a metric that tells what percentage of time did the team actually win

from sklearn.metrics import precision_score
```

```python
precision_score(test["target"], preds)
```

```python
#Split matches df up by team for computing team averages for team perfomance. Creating one df for every squad in data

grouped_matches = matches.groupby("team")
```

```python
#All matches with Manchester City

group = grouped_matches.get_group("Manchester City").sort_values("date")
```

```python
group
```

```python
#Taking a group in, set of columns what we want to compute rolling averages for and then it'll take a set of new columns that we want to assign the rolling averages

def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
#Taking columns for computing rolling averages 

cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]

#Adding rolling part to the column
new_cols = [f"{c}_rolling" for c in cols]
```

```python
#Calling rolling_averages for a single group

rolling_averages(group, cols, new_cols)
```

```python
#Applying on function to each team essentialy

matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
```

```python
#Drop extra index level

matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
#Assign values from zero to 1316 (we did unique values for each index)

matches_rolling.index = range(matches_rolling.shape[0])
```

```python
#Creating a function what will help to make predictions easier (it includes all written code but in one function)

def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
#Adding new predictior what we created

combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

```python
#Looking for an error. It is better now 

error
```

```python
#We can see where misprediction exactly are 

combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
```

```python
#By using this class we can avoid any misunderstandings in names. Doing it by pandas map method

class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
#Using created class on dataframe

combined["new_team"] = combined["team"].map(mapping)
```

```python
#Merging DataFrame with itself. I want to see if predictions match up on both sides 

merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
merged
```

```python
#Looking for all rows where one team predicted to win and other predicted to lose, it helps algorithm to be more confident in

merged[(merged["predicted_x"] == 1) & (merged["predicted_y"] ==0)]["actual_x"].value_counts()
```
